```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
def taylor_series(func, x0, n):
    x = sympy.Symbol('x')
    series = func.series(x, x0=x0, n=n).removeO()
    return sympy.lambdify(x, series)
```

```python
# Define the function to approximate
x = sympy.Symbol('x')
func = sympy.exp(x)
test=0.5
# Approximate the function at x0 = 0
x0 = 0
n = 10
approx_func = taylor_series(func, x0, n)
error=abs(approx_func(test) -  np.exp(test))
# Print the approximation
print('After %i iterations, the error is %.15f and the result is %.15f'%(n,error,approx_func(test)))
```

```python
# Define the function to approximate
x = sympy.Symbol('x')
func = sympy.sin(x)
value=1
# Approximate the function at x0 = 0
x0 = 0
x1=1.1
# Print the approximation
print("For x = %f" %value)
for n in range(1,10,1):
    y = taylor_series(func, x0, n)
    z=taylor_series(func, x1, n)
    error=abs(y(value) -  np.sin(value))
    error1=abs(z(value) -  np.sin(value))
    print("n=%i    x0=%.1f    error=%.15f       n=%i    x1=%.1f    error=%.15f" %(n,x0,error,n,x1,error1))
x = np.linspace(-2*np.pi,2*np.pi , 100)

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10,8))
y = taylor_series(func, 0, 5)
plt.plot(x,y(x), label=str(5)+str('th order'))
y = taylor_series(func, 0, 10)
plt.plot(x,y(x), label=str(10)+str('th order'))
y = taylor_series(func, 0, 20)
plt.plot(x,y(x), label=str(20)+str('th order'))
plt.plot(x,np.sin(x), label=str('Sin(x)'),color='black')

plt.grid()
plt.title('Taylor Series Approximations of Sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.linspace(-1,1 , 100)
y = np.zeros(len(x))
z = np.linspace(-1,-0.75 , 100)
labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']
func=sympy.asin(sympy.Symbol('x'))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
for n in range(10,51,10):
    y = taylor_series(func, 0, n)
    ax1.plot(x,y(x), label=str(n)+str('th order'))
    ax2.plot(z,y(z), label=str(n)+str('th order'))

ax1.plot(x, np.arcsin(x), 'k', label = 'ArcSin(x)')
ax1.grid()
ax2.plot(z, np.arcsin(z), 'k', label = 'ArcSin(x)')
ax2.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.legend()
ax2.legend()
ax1.title.set_text('Taylor Series Approximations of ArcSin(x)')
ax2.title.set_text('Zoomed Taylor Series Approximations of ArcSin(x)')
plt.show()
```
